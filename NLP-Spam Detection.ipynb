```python
#Importing necessary libraries
import pandas as pd
import numpy as np

#Creating Document term matrix
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

#Splitting the data
from sklearn.model_selection import train_test_split

#Model for prediction
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB

#Model Validation metrics
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix

#Visualizations
import seaborn as sns
import matplotlib.pyplot as plt

#Visualizing the wordcloud
from wordcloud import WordCloud
```

```python
#Reading the dataset
df=pd.read_table('sms.tsv',header=None,names=['label','message'])
```

```python
#First 5 observations
df.head()
```

```python
#Shape
print(f'No of rows : {df.shape[0]}')
print(f'No of columns : {df.shape[1]}')
```

```python
#Labelling Ham as 0,Spam as 1.
df['label']=df['label'].map({'ham':0,'spam':1})
```

```python
#Sanitary check
df.head()
```

```python
x=df['message']
y=df['label']
```

```python
#Splitting the data
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.80,random_state=10)
```

```python
#No of rows and columns after split 
print(f'No of rows & columns : {xtrain.shape}')
print(f'No of rows & columns : {xtest.shape}')
print(f'No of rows & columns : {ytrain.shape}')
print(f'No of rows & columns : {ytest.shape}')
```

```python
#Calling the CountVectorizer function
cv=CountVectorizer()
```

```python
#Fitting the countvetorizer on the train data
cv.fit(xtrain)
```

## Creating a Document Term Matrix.

```python
#Transforming the data into number counts.
xtrain_dtm=cv.transform(xtrain)
xtrain_dtm
```

## Model Building : Logistic Regression

```python
#Calling the function
lr=LogisticRegression()
```

```python
#Fitting the model
model_lr=lr.fit(xtrain_dtm,ytrain)
```

```python
#Defining a function for metrics
def validation(test,pred):
    print(f'Accuracy Score : {accuracy_score(test,pred)}')
    print(f'Precision Score : {precision_score(test,pred)}')
    print(f'Recall Score : {recall_score(test,pred)}')
    print(f'F1 Score : {f1_score(test,pred)}')
    print(f'Confusion Matrix : {confusion_matrix(test,pred)}')
    
```

```python
#Applying the function on the train dataset.
validation(ytrain,lr.predict(xtrain_dtm))
```

* Observations :
    * The model shows good performance in the train dataset.
    

```python
#Creating a document term matrix for the test set
xtest_dtm=cv.transform(xtest)
```

```python
#Applying the function on the train dataset.
validation(ytest,lr.predict(xtest_dtm))
```

* Observations :
    * The model good performance on the test dataset.
    * The Recall score has reduced but not significantly.
    * The F1 score has reduced but not significantly.

## Model Building : Gaussian Naive Bayes

```python
#Calling the function
nb=GaussianNB()
```

```python
nb.fit(xtrain_dtm.toarray(),ytrain)
```

```python
#Applying the function on the train dataset.
validation(ytrain,nb.predict(xtrain_dtm.toarray()))
```

* Observations :
    * Accuracy score has reduced.
    * The precision score has reduced.
    * The recall score has improved,which is a better metric for us because we are classifying spam(positive).

```python
#Applying the function on the train dataset.
validation(ytest,nb.predict(xtest_dtm.toarray()))
```

* Observations :
    * The performance on the test dataset has reduced.
